```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import sympy as sp
```

```python
def funcion_distribucion(x):
    if x<0:
        return 0
    elif x<1:
        return 0.5
    else:
        return 1
    
```

```python
probabilidad_cara=funcion_distribucion(1)-funcion_distribucion(0)
probabilidad_cruz=funcion_distribucion(0)-funcion_distribucion(-1)

print(probabilidad_cara)
print(probabilidad_cruz)
```

```python
x_vals=[-1,0,1,2]
y_vals=[funcion_distribucion(x) for x in x_vals]


plt.step(x_vals,y_vals,where='post',label='Funcion de distribucion')
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Funcion distribucion de la variable aleatoria')
plt.legend()
plt.grid(True)
plt.show()
```
